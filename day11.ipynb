````python
# ç¬¬ä¸€æ¬¡å°è¯•ï¼Œé—®é¢˜å¦‚ä¸‹
from collections import deque
from typing import List

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        deq = deque()
        max_list = []
        # pre compute initial deq
        for i in range(len(nums)-k):
            if i < k:
                deq.append(nums[i])
            else:
                max_list.append(max(deq))
                deq.append(nums[i])
                deq.popleft()
                
        return max_list
````

### æ ¸å¿ƒæ€æƒ³æ˜¯ æ¯æ¬¡åŠ å…¥å€¼çš„åï¼ŒæŠŠæœ€å¤§å€¼å‰é¢çš„æ‰€æœ‰å€¼éƒ½å¼¹å‡º
### è¿™æ ·popleftæ—¶å¼¹å‡ºçš„å°±ä¸€å®šæ˜¯æœ€å¤§å€¼

# å¦‚æœåŠ å…¥çš„æ–°å€¼å°±æ˜¯æœ€å¤§å€¼ï¼Œé‚£ä¹ˆè¯´æ˜å·¦è¾¹çš„å…ƒç´ ï¼ˆå½“å‰æœ€å¤§å€¼ï¼‰éœ€è¦å¼¹å‡ºäº†

````python
from collections import deque
from typing import List

class  MyQueue:
    def __init__(self):
        self.queue = deque()

    #æ¯æ¬¡å¼¹å‡ºçš„æ—¶å€™ï¼Œæ¯”è¾ƒå½“å‰è¦å¼¹å‡ºçš„æ•°å€¼æ˜¯å¦ç­‰äºé˜Ÿåˆ—å‡ºå£å…ƒç´ çš„æ•°å€¼ï¼Œå¦‚æœç›¸ç­‰åˆ™å¼¹å‡ºã€‚
    #åŒæ—¶popä¹‹å‰åˆ¤æ–­é˜Ÿåˆ—å½“å‰æ˜¯å¦ä¸ºç©ºã€‚
    def pop(self, value):
        if self.queue and value == self.queue[0]:
            self.queue.popleft()

    #å¦‚æœpushçš„æ•°å€¼å¤§äºå…¥å£å…ƒç´ çš„æ•°å€¼ï¼Œé‚£ä¹ˆå°±å°†é˜Ÿåˆ—åç«¯çš„æ•°å€¼å¼¹å‡ºï¼Œç›´åˆ°pushçš„æ•°å€¼å°äºç­‰äºé˜Ÿåˆ—å…¥å£å…ƒç´ çš„æ•°å€¼ä¸ºæ­¢ã€‚
    #è¿™æ ·å°±ä¿æŒäº†é˜Ÿåˆ—é‡Œçš„æ•°å€¼æ˜¯å•è°ƒä»å¤§åˆ°å°çš„äº†ã€‚
    def push(self, value):
        while self.queue and value > self.queue[-1]:
            self.queue.pop()
        self.queue.append(value)
    
    def front(self):
        return self.queue[0]

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        que = MyQueue()
        max_list = []
        # pre compute initial deq
        for i in range(len(nums)):
            if i < k - 1:
                que.push(nums[i])
            elif i == k-1:
                que.push(nums[i])
                max_list.append(que.front())
            else:
                que.pop(nums[i - k]) #æ»‘åŠ¨çª—å£ç§»é™¤æœ€å‰é¢å…ƒç´ 
                que.push(nums[i]) #æ»‘åŠ¨çª—å£å‰åŠ å…¥æœ€åé¢çš„å…ƒç´ 
                max_list.append(que.front())

        return max_list
````

````python
# ç¬¬ä¸€æ¬¡å°è¯•ï¼Œç”¨ deque æ¨¡æ‹Ÿ min heap

from collections import deque
from typing import List
from collections import Counter


# å…ˆç»Ÿè®¡ä¸€é å…ƒç´ ï¼šcountï¼Œç”¨counter
# ç„¶åç”¨ min heap ç»´æŠ¤ k ä¸ªå…ƒç´ ï¼Œ æ¯æ¬¡åŠ è¿›ä¸€ä¸ªå…ƒç´ ï¼Œpopå‡ºæœ€å°çš„ï¼ˆå †é¡¶ï¼‰

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        freq = Counter(nums)
        min_heap = MinHeap(k)

        for key, value in freq.items():
            min_heap.push(value)

        return min_heap.que


class MinHeap:
    def __init__(self, k):
        self.que = deque()
        self.k = k
    def push(self, num):
        if len(self.que) == 0:
            self.que.append(num)
        elif len(self.que) < self.k:
            if num >= self.que[0]:
                self.que.appendleft(num)
            else:
                self.que.append(num)
        # else, evaluate the input num and only keep the highest k values
        else:
            if num >= self.que[0]:
                self.que.appendleft(num)
                self.que.pop()
            elif num > self.que[-1]:
                self.que.pop()
                self.que.append(num)


        
````

````python
# ç¬¬äºŒæ¬¡å°è¯•ï¼Œç”¨ deque æ¨¡æ‹Ÿ min heapï¼Œ ä¿®æ­£æœ€å¼€å§‹åº”è¯¥æ˜¯ä¿ç•™æœ€å°å€¼

# è¿™ä¸ªæ–¹æ³•åœ¨æ’å…¥æ—¶éœ€è¦éå†ä¸€équeueï¼Œæ•ˆç‡æ¯”è¾ƒä½

from collections import deque
from typing import List
from collections import Counter


# å…ˆç»Ÿè®¡ä¸€é å…ƒç´ ï¼šcountï¼Œç”¨counter
# ç„¶åç”¨ min heap ç»´æŠ¤ k ä¸ªå…ƒç´ ï¼Œ æ¯æ¬¡åŠ è¿›ä¸€ä¸ªå…ƒç´ ï¼Œpopå‡ºæœ€å°çš„ï¼ˆå †é¡¶ï¼‰

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        freq = Counter(nums)
        min_heap = MinHeap(k)

        for key, value in freq.items():
            min_heap.push((value,key))

        return [num for _, num in min_heap.que]



class MinHeap:
    def __init__(self, k):
        self.que = deque()  # ä½¿ç”¨ deque å­˜å‚¨å †
        self.k = k

    def push(self, freq_elem):
        """freq_elem æ˜¯ä¸€ä¸ªå…ƒç»„ (é¢‘ç‡, å…ƒç´ )"""
        freq, num = freq_elem

        # å¦‚æœå †æœªæ»¡ï¼Œç›´æ¥æ’å…¥åˆ°åˆé€‚çš„ä½ç½®
        if len(self.que) < self.k:
            self._insert(freq_elem)
        # å¦‚æœå †å·²æ»¡ï¼Œåªåœ¨å½“å‰é¢‘ç‡å¤§äºæœ€å°å€¼æ—¶æ’å…¥
        elif freq > self.que[0][0]:
            self.que.popleft()  # ç§»é™¤å †é¡¶æœ€å°å€¼
            self._insert(freq_elem)

    def _insert(self, freq_elem):
        """æŒ‰å‡åºæ’å…¥ freq_elem (é¢‘ç‡, å…ƒç´ )"""
        freq, _ = freq_elem

        # éå†æŸ¥æ‰¾æ’å…¥ä½ç½®
        for i in range(len(self.que)):
            if freq < self.que[i][0]:  # æ¯”è¾ƒé¢‘ç‡, æ‰¾åˆ°æ¯”å½“å‰å€¼å¤§çš„ä½ç½®æ’å…¥
                self.que.insert(i, freq_elem)
                return
        # å¦‚æœæœªæ‰¾åˆ°æ’å…¥ä½ç½®ï¼Œç›´æ¥æ·»åŠ åˆ°é˜Ÿå°¾
        self.que.append(freq_elem)
````

### **å¤æ‚åº¦åˆ†æ**

  - ç»Ÿè®¡é¢‘ç‡ï¼š$O(n)$ï¼Œå…¶ä¸­ $n$ æ˜¯æ•°ç»„é•¿åº¦ã€‚
  - å †æ“ä½œï¼šæ¯ä¸ªå…ƒç´ æ’å…¥å †çš„æ—¶é—´å¤æ‚åº¦ä¸º $O(k)$ï¼Œæœ€å¤šæ’å…¥ $m$ ä¸ªå…ƒç´ ï¼ˆ$m$ æ˜¯æ•°ç»„ä¸­ä¸åŒçš„å…ƒç´ æ•°ï¼‰ã€‚
  - æ€»å¤æ‚åº¦ï¼š$O(n + m \cdot k)$ã€‚

- **ç©ºé—´å¤æ‚åº¦**ï¼š
  - ä½¿ç”¨ `deque` å­˜å‚¨å †ï¼š$O(k)$ã€‚
  - ä½¿ç”¨ `Counter` å­˜å‚¨é¢‘ç‡ï¼š$O(m)$ã€‚
  - æ€»ç©ºé—´å¤æ‚åº¦ï¼š$O(m + k)$ã€‚

````python
# å­¦ä¹ ç­”æ¡ˆï¼Œç”¨heapq

#æ—¶é—´å¤æ‚åº¦ï¼šO(nlogk)
#ç©ºé—´å¤æ‚åº¦ï¼šO(n)
import heapq
class Solution:
    
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        #è¦ç»Ÿè®¡å…ƒç´ å‡ºç°é¢‘ç‡
        map_ = {} #nums[i]:å¯¹åº”å‡ºç°çš„æ¬¡æ•°
        for i in range(len(nums)):
            map_[nums[i]] = map_.get(nums[i], 0) + 1
        
        #å¯¹é¢‘ç‡æ’åº
        #å®šä¹‰ä¸€ä¸ªå°é¡¶å †ï¼Œå¤§å°ä¸ºk
        pri_que = [] #å°é¡¶å †
        
        #ç”¨å›ºå®šå¤§å°ä¸ºkçš„å°é¡¶å †ï¼Œæ‰«ææ‰€æœ‰é¢‘ç‡çš„æ•°å€¼
        for key, freq in map_.items():
            heapq.heappush(pri_que, (freq, key))
            if len(pri_que) > k: #å¦‚æœå †çš„å¤§å°å¤§äºäº†Kï¼Œåˆ™é˜Ÿåˆ—å¼¹å‡ºï¼Œä¿è¯å †çš„å¤§å°ä¸€ç›´ä¸ºk
                heapq.heappop(pri_que)
        
        #æ‰¾å‡ºå‰Kä¸ªé«˜é¢‘å…ƒç´ ï¼Œå› ä¸ºå°é¡¶å †å…ˆå¼¹å‡ºçš„æ˜¯æœ€å°çš„ï¼Œæ‰€ä»¥å€’åºæ¥è¾“å‡ºåˆ°æ•°ç»„
        result = [0] * k
        for i in range(k-1, -1, -1):
            result[i] = heapq.heappop(pri_que)[1]
        return result
````

````python
# ç¬¬ä¸‰æ¬¡å°è¯•ï¼Œç”¨heapqç›´æ¥å®ç°æœ€å°å †ï¼Œç›´æ¥å­˜å‚¨ (value,key) pair, å¹¶æŒ‰ç…§ (value,key) ä¸­valueæ’åº
# 23ms -> 7ms, æ•ˆç‡æå‡å¾ˆå¤š

import heapq
from collections import Counter
from typing import List
# heapä¹Ÿå¯ä»¥å…¼å®¹tupleï¼Œä½†æ˜¯é»˜è®¤æŒ‰ç…§ tuple çš„ index 0 æ’åºï¼Œ heapqæ²¡æœ‰ç›´æ¥æ”¯æŒå¯¹å…ƒç»„ç¬¬äºŒä¸ªå…ƒç´ æ’åºçš„é€‰é¡¹
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        
        freq = Counter(nums)
        min_heap = []

        for key, value in freq.items():
            heapq.heappush(min_heap, (value,key)) # æ³¨æ„è¿™é‡Œé¢‘ç‡åœ¨å‰
            if len(min_heap) > k:
                heapq.heappop(min_heap)
        
        return [key for _, key in min_heap]
````
